```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('vectorized_responses.pkl', 'rb') as f:
    X = pickle.load(f)
```

```python
actions = X[:,:,-1,:].argmax(axis=2)
means = X[:,:,0,:]
var = X[:,:,1,:]
mes = X[:,:,2,:]
f = X[:,:,4,:]
```

```python
def plot_scores(idx):
    for i in range(X.shape[1]):
        a = actions[idx][:i]
        fi = f[idx][i][a]
        m = means[idx][i]
        v = var[idx][i]
        me = mes[idx][i]
        std = np.sqrt(v)
        lower = m - std
        upper = m + std

        plt.plot(a, fi,'bo');
        plt.plot(m);
        plt.fill_between(np.arange(len(m)), lower, upper, alpha=.5)
        plt.plot(me*400)
        print (me.argmax())
        plt.show()
```

```python
plot_scores(0)
```

```python
plot_scores(4)
```

```python
plot_scores(5)
```
